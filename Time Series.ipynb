```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Topsus/Data Topsus.csv', sep=';') #Membaca file dengan memisahkan berdasarkan ;
```

# Preprocessing Data

```python
df.info()
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date']) #Menggubah ke format date
```

```python
#Mengubah type data ke float
df['Sales'] = df['Sales'].str.replace(',', '.').astype(float)
df['Profit'] = df['Profit'].str.replace(',', '.').astype(float)
```

```python
df.info()
```

```python
#Group berdasarkan sales
sales_by_ship_date = df.groupby("Ship Date")["Sales"].sum()
```

```python
print(sales_by_ship_date)
```

#Use Data

```python
#Mengubah nama data dan memilih data yang digunakan
df = df[['Ship Date', 'Sales']].rename({'Ship Date':'shipdate', 'Sales':'sales'}, axis='columns')
```

```python
df.head()
```

```python
df.tail()
```

```python
date_df=df.set_index('shipdate') #Mengubah ke index
```

```python
date_df_monthly=date_df.resample('M').sum() #urut berdasarkan month dan jumlahkan sales setiap bulan
```

```python
date_df_monthly.head(20)
```

```python
date_df_monthly.shape
```

```python
date_df_monthly.plot();
```

# Time Series Forecasting

```python
from pycaret.time_series import *
```

```python
#Experiment Forecasting
exp = setup(date_df_monthly, fh=7, fold=3, session_id=123)
```

```python
check_stats()
```

#Plot Data

```python
exp.plot_model(plot = 'decomp_stl')
```

```python
exp.plot_model(plot='diagnostics')
```

#Build Model

```python
best = compare_models()
```

```python
best
```

#Plot Best Model

```python
exp.plot_model(best, plot='insample')
```

```python
exp.plot_model(best, plot='residuals')
```

#Analyze Model

```python
#Peramalan untuk kebutuhan masa depan
#Misal 24 bulan berikutnya
exp.plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 24})
```

```python
exp.plot_model(best, plot='diagnostics')
```

```python
final_best = exp.finalize_model(best)
exp.predict_model(best, fh = 24)
```

#Save Model

```python
save_model(final_best, 'my_final_best_model')
```

```python
loaded_model = load_model('my_final_best_model')
print(loaded_model)
```
